### Data Cleaning 

```python
#Eliminando columnas 

data.drop(columns = ['Unnamed: 8', 'Unnamed: 9'], inplace = True)
```

Al tener todos nuestros valores sin valores nulos, podemos arrancar con el análisis de nuestras bases

```python
data = data.drop(data.index[1528])
```

```python
data['reclamo_cantidad'] = pd.to_numeric(data['reclamo_cantidad'])
```

```python
data.head()
```

```python
data['reclamo_cantidad'].sum()
```

```python
data.groupby('reclamo_ano')['reclamo_cantidad'].sum().plot.bar(colormap = "Dark2")
plt.title('Cantidad de reclamos solucionados por año', fontsize = 20)
plt.xlabel('Años')
plt.ylabel('Cantidad')
plt.show()
```

```python
data.groupby('reclamo_ano')['reclamo_cantidad'].sum()
```

```python
data.groupby('reclamo_ano')['reclamo_cantidad'].max()
```

```python
data[data['reclamo_cantidad'] == 281]
```

```python
data.groupby('reclamo_mes_nombre')['reclamo_cantidad'].sum().sort_values(ascending = False).plot.bar(colormap='Spectral')
plt.title('Cantidad de reclamos por mes',fontsize = 20)
plt.ylabel('Meses', fontsize = 15)
plt.xlabel('Cantidad', fontsize = 15)
plt.show();
                                                              
```

```python
data.groupby('reclamo_mes_nombre')['reclamo_cantidad'].mean().sort_values(ascending = False).plot.bar(colormap = "Spectral")
plt.title('Media de reclamos por mes',fontsize = 20)
plt.ylabel('Meses', fontsize = 15)
plt.xlabel('Cantidad de la media', fontsize = 15)
plt.show();
```

```python
data.groupby('reclamo_mes_nombre')['reclamo_cantidad'].max().sort_values(ascending = False).plot.bar(colormap = "Spectral")
plt.title('Máxima de reclamos por mes',fontsize = 20)
plt.ylabel('Meses', fontsize = 15)
plt.xlabel('Máximos', fontsize = 15)
plt.show();
```

```python
#Se reemplazan los valores telefónicos sin acento para que no parezcan otra variable

data['reclamo_origen'] = data['reclamo_origen'].str.replace('Telefonico', 'Telefónico')
```

```python
data.groupby('reclamo_origen')['reclamo_cantidad'].sum().sort_values(ascending = False).plot.bar(colormap = "autumn")
plt.title('Medios de reclamos más usados',fontsize = 20)
plt.ylabel('Medios usados', fontsize = 15)
plt.xlabel('Cantidad', fontsize = 15)
plt.show();
```

```python
data['reclamo_tipo'] = data['reclamo_tipo'].str.replace('I   - Facturación y Gestión de Deudas', 
                                                        'I   - Facturacion y Gestion de Deudas')
```

```python
data.groupby('reclamo_tipo')['reclamo_cantidad'].sum().sort_values(ascending = False).plot.bar(color = 'blue')
plt.title('Tipos de reclamos más populares')
plt.xlabel('Reclamos')
plt.ylabel('Cantidad')
plt.show()
```
